# 2.3.3 Indexing Data

```python
a = {'a':[1,2,3,4],'b':[5,6,7,8],'c':[9,10,11,12],'d':[13,14,15,16]} #create a dictionary of lists
df = pd.DataFrame(a) #create a data frame (a Pandas tabular data structure) from the dictionary of lists
df
```

Note: columns and rows begin with index 0

```python
df['c'][1] #select the value in column 'c', row 1
```

```python
df.iloc[[0,2],[1,3]] #select rows 0 and 2, and columns 1 and 3
```

Note: iloc is primarily integer based

```python
df[0:3][['b','c','d']] #select the first three rows, and last three columns
```

```python
df[:2] #select first two rows
```

```python
df[['a','b']] # select first two columns
```

```python
df[:1] #select the first row
```

```python
df.drop(df.index[[0,2]]) #drop row 0 and 2
```

```python
df.drop(df.index[[0,2]]).drop(['a','c','d'], axis=1)
```

```python
df.shape #dataframe dimension (number rows, number columns)
```
