# 3.6.2 Simple Linear Regression

```python
# Import data set and list column names

import pandas as pd
import statsmodels.api as sm

boston = sm.datasets.get_rdataset("Boston", "MASS", cache=True).data
print boston.dtypes
```

Note: For a list of available R data sets go to http://vincentarelbundock.github.io/Rdatasets/

```python
boston[:3] #first three rows
```

```python
# Scatterplot and descriptive statistics of lstat (percent of housholds with low socioeconomic status) and medv (median house value)

%matplotlib inline
import matplotlib.pyplot as plt

x = boston.lstat
y = boston.medv 
plt.plot(x, y, 'bo')

boston[['lstat','medv']].describe() 
```

```python
import numpy as np
import statsmodels.formula.api as smf

# Fit regression model 
results = smf.ols('medv ~ lstat', data=boston).fit() #lowercase models (in this case: ols) accept formula and df arguements

# Inspect the results
print results.summary()
```

Note: statsmodels is a Python module that provides classes and functions for the estimation of many different statistical models, as well as for conducting statistical tests, and statistical data exploration.

```python
#out of sample prediction

Xnew = {'lstat':[10,20,30]}
Xnew = pd.DataFrame(Xnew)
ynewpred = results.predict(Xnew)
ynewpred
```

```python
#how to output confidence/prediction interval for out of sample predictions?
```

```python
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std

prstd, iv_l, iv_u = wls_prediction_std(results) #standard error of prediction, and 95% confidence bands

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x, y, 'o', label="data")
ax.plot(x, results.fittedvalues, 'r--.', label="OLS")
ax.plot(Xnew, ynewpred, 'ys', label='o-o-s')
ax.plot(x, iv_u, 'r--')
ax.plot(x, iv_l, 'r--')
ax.legend(loc='best')
```

# 3.6.4 Interaction Terms

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Import data
boston = sm.datasets.get_rdataset("Boston", "MASS", cache=True).data

# OLS with interaction terms (see note)
results_int = smf.ols('medv ~ lstat*age', data=boston).fit()

#Inspect the results
print results_int.summary()
```

Note: ”:” adds a new column to the design matrix with the product of the other two columns. “*” will also include the individual columns that were multiplied together
