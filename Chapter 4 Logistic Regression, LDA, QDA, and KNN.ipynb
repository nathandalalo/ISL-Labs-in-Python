# 4.6.2 Logistic Regression

Resources: 

http://statsmodels.sourceforge.net/stable/examples/notebooks/generated/discrete_choice_example.html

http://nbviewer.ipython.org/github/carljv/Will_it_Python/blob/master/ARM/ch5/arsenic_wells_switching.ipynb

```python
import pandas as pd
import numpy as np

# Import data
smarket = pd.read_csv("C:\Users\Nathan\Dropbox\Statistical Learning Data Sets\Smarket.csv")
smarket = smarket.drop("Unnamed: 0",1)
```

```python
# Re-code 'Directions' for logistic regression
# Up = 1, Down = 0

smarket['Direction_y'] = (smarket.Direction=='Up').astype(int)

smarket.head(5)
```

```python
import statsmodels.formula.api as smf

# Fit regression model 
results = smf.logit('Direction_y ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume', data=smarket).fit()

# Inspect the results
print results.summary()
```

```python
# P(Y=1|X) for the first 10 observations in the training dataset 

print results.predict(smarket[:10])
```

```python
# Create confusion matrix, model (training) error rate, and null error rate
# Note: the null error rate comes from a model that just classifies everything as whatever the most prevalent response is

print results.pred_table()
print 'Model Error rate: {0: 3.0%}'.format(1 - np.diag(results.pred_table()).sum() / results.pred_table().sum())
print 'Null Error Rate: {0: 3.0%}'.format(1 - smarket['Direction_y'].mean())
```

```python
# Training set (Year: 2001-2004)
# Test set (Year: 2005)

smarket_train = smarket[smarket.Year < 2005]
smarket_test = smarket[smarket.Year == 2005]

print smarket_train.shape
print smarket_test.shape
```

```python
# Fit regression model with data from 2001-2004
results_train = smf.logit('Direction_y ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume', data=smarket_train).fit()

# Inspect the results
print results_train.summary()
```

```python
# Create confusion matrix, model (training) error rate, and null error rate
# Note: the null error rate comes from a model that just classifies everything as whatever the most prevalent response is

pred = np.array(results_train.predict(smarket_test) > 0.5, dtype=float)
c_matrix_test = np.histogram2d(smarket_test.Direction_y, pred, bins=2)[0]
print c_matrix_test

print 'Model Error rate: {0: 3.0%}'.format(1 - np.diag(c_matrix_test).sum() / c_matrix_test.sum())
print 'Null Error Rate: {0: 3.0%}'.format(1 - smarket_test['Direction_y'].mean())
```

```python
# Fit regression model with data from 2001-2004 using only Lag1 and Lag2
results_train2 = smf.logit('Direction_y ~ Lag1 + Lag2', data=smarket_train).fit()

# Inspect the results
print results_train2.summary()
```

```python
# Create confusion matrix, model (training) error rate, and null error rate
# Note: the null error rate comes from a model that just classifies everything as whatever the most prevalent response is

pred2 = np.array(results_train2.predict(smarket_test) > 0.5, dtype=float)
c_matrix_test2 = np.histogram2d(smarket_test.Direction_y, pred2, bins=2)[0]
print c_matrix_test2

print 'Model Error rate: {0: 3.0%}'.format(1 - np.diag(c_matrix_test2).sum() / c_matrix_test2.sum())
print 'Null Error Rate: {0: 3.0%}'.format(1 - smarket_test['Direction_y'].mean())
```

```python
# Predict the returns when Lag1=1.2 and Lag2=1.1
senario1 = pd.DataFrame({'Lag1':[1.2], 'Lag2':[1.1]})
print results_train2.predict(senario1)

# Predict the returns when Lag1=1.5 and Lag2=-0.8
senario2 = pd.DataFrame({'Lag1':[1.5], 'Lag2':[-0.8]})
print results_train2.predict(senario2)
```
